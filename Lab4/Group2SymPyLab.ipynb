## SymPy’s Symbolic and Numercical Complex Evaluations


a = 9 + i a2^2, b = b1^2 + 3 i, c = 2 c1 + i c2

image.png

```
from sympy import symbols, I, N, pi, expand
a1, b1, c1, c2 = symbols("a1 b1 c1 c2", real=True)  
a = 9 + I*a1**2
b = b1**2 + 3*I
c = 2*c1 + I*c2

print(a*b*c)
print(expand(a*b*c))
print(expand((a*b)*c))
print(expand(a*(b*c)))

w=N(1/(pi + I), 20)
print('w=',w)
```
